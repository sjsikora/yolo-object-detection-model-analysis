```python
from google.colab import drive, runtime
drive.mount('/content/drive')

from ultralytics import YOLO
from pathlib import Path
from PIL import Image
import shutil

DST_DIR = "/content/drive/MyDrive/CMPE401/YOLO/data"
yaml_path = Path(DST_DIR) / "visdrone.yaml"
```

# Training the Base Model

```python
model = YOLO("yolo11s.pt")

model.train(
    data=str(yaml_path),
    epochs=50,
    imgsz=640,
    batch=32,           # sweet spot for VisDrone small objects
    device=0,           # explicitly use the A100
    workers=4,          # Colab typically has 2-4 CPU cores
    name="visdrone_yolo11s_baseline",
    project="/content/drive/MyDrive/CMPE401/YOLO/runs",

    # Optimizer
    optimizer="AdamW",  # more stable than default SGD for dense scenes
    lr0=0.001,          # initial learning rate
    lrf=0.01,           # final lr = lr0 * lrf
    warmup_epochs=3,    # gradual warmup, helps with dense datasets

    # Regularization
    weight_decay=0.0005,
    dropout=0.0,        # keep off for baseline

    # Augmentation (YOLO defaults are good, these tune for small objects)
    mosaic=1.0,         # mosaic augmentation — very helpful for VisDrone
    mixup=0.1,          # mild mixup
    copy_paste=0.1,     # helps small object recall

    # Saving & logging
    patience=15,        # early stopping — more generous given your GPU
    save=True,
    save_period=10,     # checkpoint every 10 epochs as backup
    plots=True,
    verbose=True,
)
```
